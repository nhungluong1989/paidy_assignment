# Check missing data

```python
total_null_records = df.isnull().sum().sort_values(ascending = False) # the number of null records
total_records = df.isnull().count().sort_values(ascending = False) # the number of records including null records
total_records_excl_null = df.count().sort_values(ascending = False) # the number of records excluding null records
percent = (total_null_records/total_records*100).sort_values(ascending = False)
```

```python
total_null_records
```

```python
percent
```

```python
pd.concat([total_null_records, percent], axis = 1, keys = ['Total_Null_Records', 'Percentage_Null_Records']).transpose()
```

- There is 19.82% missing data in MonthlyIncome
- There is 2.62% missing data in NumberOfDependents
- For other features, there is no missing data

# Handle missing data
- Apply mean for missing data in MonthlyIncome
- Apply mode for missing data in NumberOfDependents

```python
df['MONTHLY_INCOME'].fillna(int(df['MONTHLY_INCOME'].mean()), inplace = True)
df
```

```python
df['MONTHLY_INCOME'].mean()
```

```python
df['TOTAL_DEPEDENTS'].fillna(df['TOTAL_DEPEDENTS'].mode()[0], inplace = True)
df
```

```python
df.isnull().sum().sort_values(ascending = False)
```

There is no missing data after handling

# Data exploration

# Features correlation

```python
plt.figure(figsize = (20,20))
plt.title('Loan Applicant Information Correlation Plot (Pearson)')
corr = df.corr()
corr
sns.heatmap(corr, xticklabels = corr.columns, yticklabels = corr.columns, linewidths = .1, cmap = 'Reds')
plt.show()
```

There are certain direct correlations between: 
- TOTAL_60_89_DAYS_PAST_DUE & TOTAL_30_59_DAYS_PAST_DUE 
- TOTAL_90_DAYS_LATE & TOTAL_30_59_DAYS_PAST_DUE 
- TOTAL_60_89_DAYS_PAST_DUE & TOTAL_90_DAYS_LATE 

The corr of these features is close to 1 means strong positive correlation. It means that when a person has been 30-59 days past due, it's is likely that he/she will be having 60-89 days past due or 90 days or more past due.

# The relationship between age and MonthlyIncome 

```python
df1 = df.loc[:, ['AGE', 'MONTHLY_INCOME']].groupby('AGE').sum().reset_index(level = 0)
df1
```

```python
plt.scatter(df1['AGE'], df1['MONTHLY_INCOME'])
plt.xlabel('AGE')
plt.ylabel('MONTHLY_INCOME')
plt.title('AGE vs MONTHLY_INCOME')
plt.show()
```

- The people between the ages of 40 and 65 have the highest monthly income, because they are of working age and have many years of experience
- Whereas the people under 22 years old get the lowest monthly income, since they are still at school without working full-time (only a few people have income from part-time job)
- And similar trend for the people over 70 years old, because they already retired with only pension.
